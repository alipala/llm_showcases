```python
import getpass
import os

os.environ["LLAMA_CLOUD_API_KEY"] = getpass.getpass("LLMA Cloud API Key:")
```

```python
from llama_parse import LlamaParse

document = LlamaParse(result_type="markdown").load_data("apple_annual_report.pdf")
```

```python
print(document)
```

```python
print(document[0].text[:1000])
```

```python
# write document content into the markdown file

file_name = "apple_annual_report.md"
with open(file_name, "w") as file:
    file.write(document[0].text)
```

```python
document_with_instruction = LlamaParse(
    result_type="markdown",
    parsing_instruction="""
    This is a part of Apple annual report
    """
).load_data("apple_annual_report.pdf")
```

```python
file_name = "apple_annual_report_with_instruction.md"
with open(file_name, "w") as file:
    file.write(document_with_instruction[0].text)
```
